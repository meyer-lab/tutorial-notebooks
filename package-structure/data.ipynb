```python
import random
import matplotlib.pyplot as plt
```

## Plotting in Notebook

Let's assume we can separate our plot generation into two steps:
1. Some computation to get the values of interest (usually model outputs)
2. Plotting those values

The next two cells go through an example of how to do this in a notebook

```python
x = range(10)
y = [random.randrange(100) for _ in x]
print(x, y);
```

```python
plt.scatter(x,y)
plt.xlabel("x value")
plt.ylabel("y value")
plt.title("title");
```

## Plotting from Python Package  

This plot generation can also be performed with a python package by moving these two steps into functions. For this example:
The data calculation step was moved into a dataCalculation() function defined in pythonPackage/data/calculations.py
The plotting step was moved into a makeFig() function defined in pythonPackage/figures/figure1.py

All necessary libraries are imported directly into those two files. So in order to make the figure here, all we have to do is import and call the makeFig() function:

```python
from pythonPackage.figures.figure1 import makeFig

makeFig()
```
